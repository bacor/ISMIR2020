# Figure S3: Majority baselines

```python
%pylab inline 
import pandas as pd
import seaborn as sns
from helpers import cm2inch
matplotlib.style.use('styles.mplstyle')
```

```python
def majority_baseline(df):
    counts = pd.value_counts(df['mode'])
    top_mode = counts.idxmax()
    top_mode_freq = counts.max() / len(df)
    return top_mode, top_mode_freq
```

```python
majority_baselines = [
    ['responsory', 'full', 'train'],
    ['responsory', 'full', 'test'],
    ['responsory', 'subset', 'train'],
    ['responsory', 'subset', 'test'],
    ['antiphon', 'full', 'train'],
    ['antiphon', 'full', 'test'],
    ['antiphon', 'subset', 'train'],
    ['antiphon', 'subset', 'test']
]
columns=[(None, 'genre'), (None, 'dataset'), (None, 'kind')]
    
for run in range(5):
    resp_train_full = pd.read_csv(f'../data/run-{run}/responsory/full/train-chants.csv')
    majority_baselines[0].extend(majority_baseline(resp_train_full))
    resp_test_full = pd.read_csv(f'../data/run-{run}/responsory/full/test-chants.csv')
    majority_baselines[1].extend(majority_baseline(resp_test_full))
    
    resp_train_subset = pd.read_csv(f'../data/run-{run}/responsory/subset/train-chants.csv')
    majority_baselines[2].extend(majority_baseline(resp_train_subset))
    resp_test_subset = pd.read_csv(f'../data/run-{run}/responsory/subset/test-chants.csv')
    majority_baselines[3].extend(majority_baseline(resp_test_subset))
    
    anti_train_full = pd.read_csv(f'../data/run-{run}/antiphon/full/train-chants.csv')
    majority_baselines[4].extend(majority_baseline(anti_train_full))    
    anti_test_full = pd.read_csv(f'../data/run-{run}/antiphon/full/test-chants.csv')
    majority_baselines[5].extend(majority_baseline(anti_test_full))
    
    anti_train_subset = pd.read_csv(f'../data/run-{run}/antiphon/subset/train-chants.csv')
    majority_baselines[6].extend(majority_baseline(anti_train_subset))
    anti_test_subset = pd.read_csv(f'../data/run-{run}/antiphon/subset/test-chants.csv')
    majority_baselines[7].extend(majority_baseline(anti_test_subset))
    
    columns.extend([(f'run {run}', 'mode'), (f'run {run}', 'frequency')])
```

```python
# Print th
cols = pd.MultiIndex.from_tuples(columns)
df = pd.DataFrame(majority_baselines, columns=cols)
df[('avg', 'frequency')] = df.iloc[:,[4, 6, 8, 10, 12]].mean(axis=1)
df
```
